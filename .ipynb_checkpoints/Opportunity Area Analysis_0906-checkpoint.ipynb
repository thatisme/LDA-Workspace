```python
# df ÌÜµÌï©
total_df = pd.concat([df0,df1,df2,df3])
total_df.reset_index(inplace = True, drop = True)
total_df
```

### 2. Satisfaction

#### 1) Í∞Å ÌÅ¥Îü¨Ïä§ÌÑ∞Ïùò Ïï°ÏÖòÏóê ÎåÄÌï¥ÏÑú Satisfaction Í≥ÑÏÇ∞

```python
total_df.cluster.unique()
```

```python
action_sents = dict()
for actor in total_df.cluster.unique() :
    actor_df = total_df[total_df['cluster']==actor]
    for i in actor_df.action_cluster.unique() :
        action_score = actor_df[actor_df['action_cluster']==i].star
        action_score = np.mean(action_score)
        action_sents['Actor{}_Action{}'.format(actor, i)] = action_score
```

```python
action_sents
```

#### 2) Satisfaction Ï†ïÍ∑úÌôî

```python
data = action_sents.values()
data = np.array(list(data)).reshape(-1, 1) # -1ÏùÄ ÌÅ¨Í∏∞Í∞Ä 1Ïù∏ Ï∞®ÏõêÏúºÎ°ú Í≥ÑÏÇ∞ÌïòÍ≤†Îã§Îäî ÏùòÎØ∏? ÎÜâ, Ï∞®ÏõêÏùÑ ÏûêÎèôÏ†ÅÏúºÎ°ú~ Ïó¨Í∏∞ÏÑ† 1Î°ú ÏûêÎèô ÌôïÏù∏
data
```

```python
scaler = MinMaxScaler(feature_range = (-10, 10))
transposed_data = scaler.fit_transform(data)
transposed_data
```

```python
score_result = transposed_data.flatten().tolist() # flatten : Ï∞®ÏõêÏùÑ ÌíÄÏñ¥Ï£ºÎäî Ï¶â Ï§ÑÏó¨Ï£ºÎäî Î©îÏÜåÎìú
score_result
```

```python
for key, new_value in zip(action_sents.keys(), score_result) : # zip : Í∞êÏã∏Ï£ºÎäî? Ìï®Ïàò
    action_sents[key] = new_value
```

```python
action_sents
```

### 3. Importance Ï†êÏàò

#### 1) Í∞Å ÌÅ¥Îü¨Ïä§ÌÑ∞Ïùò Ïï°ÏÖòÏóê ÎåÄÌï¥ÏÑú Importance(Ï†ÑÏ≤¥ ÌÜ†ÌîΩÏùò ÎπÑÏ§ë) Ï†êÏàò Í≥ÑÏÇ∞ 

```python
# action Í∞úÏàò
importance_check = []
for actor, action in zip(total_df['cluster'], total_df['action_cluster']) :
    importance_check.append('Actor'+ str(actor) + '_' + 'Action' + str(action))

# importance_check
```

```python
frequency = Counter(importance_check)
frequency
```

```python
# Ï§ëÏöîÎèÑ ÏÇ¨Ï†Ñ ÎßåÎì§Í∏∞
importance_dict = dict()

# Ï†ÑÏ≤¥ Ìï©
total_count = sum(frequency.values())

# ÎπàÎèÑÏàò / Ï†ÑÏ≤¥Í∞í * 100
for item, value in frequency.items() :
    importance = (value/total_count) * 100
    importance_dict[item] = importance
```

#### 2) Importance Ï†êÏàò Ï†ïÍ∑úÌôî(0~10)

```python
data = importance_dict.values()

# array ÌòïÌÉúÎ°ú Î≥ÄÌôò
data = np.array(list(data)).reshape(-1,1)
```

```python
scaler = MinMaxScaler(feature_range = (0,10))
transformed_data = scaler.fit_transform(data)
transformed_data
```

```python
score_result = transformed_data.flatten().tolist()
score_result = [round(i,4) for i in score_result] # Í∑∏ÎÉ• round(score_result,4)Îäî ÏïàÎêòÎÇò?
```

```python
# value Í∞í ÍµêÏ≤¥
for key, new_value in zip(importance_dict.keys(), score_result) :
    importance_dict[key] = new_value
importance_dict
```

### 4. Opportunity

#### 1) Opportunity score
- Opportunity = Importance + Max(Importance-Satisfaction, 0)

```python
def opportunity_score(satisfaction, importance) :
    result = importance + max(importance - satisfaction, 0)
    return result
```

```python
# Í∞êÏÑ±Ï†êÏàò / opportunity_score Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÌôî
opportunity_df = pd.DataFrame(action_sents.items(), columns = ['Action', 'satisfaction'])
opportunity_df
```

```python
opportunity_df['importance'] = importance_dict.values()
opportunity_df
```

```python
opportunity_list = []
for i,j in zip(opportunity_df['satisfaction'], opportunity_df['importance']) :
    score_result = opportunity_score(i,j)
    opportunity_list.append(score_result)

opportunity_df['opportunity_score'] = opportunity_list
opportunity_df
```

```python
# Action Ï†ïÎ†¨
opportunity_df = opportunity_df.sort_values(by = 'Action')
opportunity_df.reset_index(drop=True,inplace=True)
opportunity_df.to_csv('./4. LDA_result/Opportunity Score.csv', encoding = 'utf-8-sig')
opportunity_df
```

#### 2. Opportunity area

```python
# Îç∞Ïù¥ÌÑ∞ ÏÑ∏ÌåÖ
satisfaction = opportunity_df.satisfaction
importance = opportunity_df.importance
actions = opportunity_df.Action
color = np.random.rand(len(actions), 3)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(importance, satisfaction, s = 35, c = color) # sÎäî Ï†êÏùò ÏÇ¨Ïù¥Ï¶à / cÎäî ÏÉâÏÉÅ

# ÎùºÎ≤®Ïù¥Î¶Ñ
plt.xlabel("Importance")
plt.ylabel("Satisfaction")

# Ï§ëÏöîÎèÑ Í∏∞Ï§ÄÏÑ†
xdata = [importance.mean(), 10]
ydata = [-10, 10]
plt.plot(xdata, ydata, 'k')

# ÎßåÏ°±ÎèÑ Í∏∞Ï§ÄÏÑ†
xdata = [0, 10]
ydata = [satisfaction.mean(), 10]
plt.plot(xdata, ydata, 'k')

# Ï†êÏóê Ïù¥Î¶Ñ Î∂ôÏù¥Í∏∞
texts = []
for i, action in enumerate(actions) :
    texts.append(plt.text(importance[i], satisfaction[i], action, fontsize=13))

# ÏúÑÏπòÏ°∞Ï†ï
adjust_text(texts, arrowprops = dict(arrowstyle = '-', color = 'grey', lw = 1)) # lwÎäî ÏÑ†Ïùò ÎëêÍªò

# ÏÉâÏÉÅ Î≤îÎ°Ä Ï∂îÍ∞Ä
for i, action in enumerate(actions):
    plt.scatter([], [], c=[color[i]], label=action, s=10)
 
plt.legend(title = 'Actions', fontsize=8, title_fontsize='10', loc='upper left', bbox_to_anchor=(1,1))

plt.show()
```
